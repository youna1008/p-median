## 1. p-median 모델 함수화 하기 ( test 끝 , 실전 )

```python
def p_median0 (p,cost,n):
    score = pd.DataFrame(p) # 최종 score df 만들어 놓기 
    score['비용']= np.nan # 비용 열 만들기
    
    for index,i in enumerate(p):
        i = list(i) # 첫번째 최소비용 만들기 
        index0 = list(range(n))
        concat_df =  cost.iloc[i,:] # 최소비용 계산 할 데이터프레임
    
        # for문을 통해 데이터 프레임 만들기 끝
        first_list = [] # 데이터 프레임 에서 최소 비용 담을 list 
        for k in concat_df.columns:
            min0 = min(concat_df[k])
            first_list.append(min0)
        
        plus_list = [] # 마이너스 값만 넣어줄 리스트 
        diff_row = cost.iloc[index0[-1]]- first_list
        diff_row = list(diff_row)
        for j in diff_row:
            if j < 0 :
                plus_list.append(j)
        plus_sum = sum(first_list) + sum(plus_list)
        score.iloc[index,n]=plus_sum
        
    return score
```
